```python
questions = ["1.how many people have diabetes and how many people doesn't have diabetes?",
             "2.people of which Gender has most diabetes?",
             "3.people of which Gender has which type of diagnosis most?",
             "4.people with which diagnosis are most pron to have diabetes?"]
questions
```

```python
#let find the answer of first question
#1.how many people have diabetes and how many people doesn't have diabetes?
#getting the values
df.Diagnosis.value_counts()
```

```python
#ploting bar chart
df.Diagnosis.value_counts().plot(kind = 'bar', color = ["orchid", "salmon"])
plt.title("diabetes")
plt.xlabel("Diagnosis")
plt.ylabel("count");
```

```python
#plotting a pie chart
df.Diagnosis.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.legend(["diabetes", "no diabetes"]);
```

```python
#now let's check how many 'male' and 'female' are in the dataset
df.Gender.value_counts()
```

```python
#plotting a pie chart
df.Gender.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.title('male female ratio')
plt.legend(['male', 'female']);
```

```python
#let find the answer of second question
#2.people of which Gender has most Diabetes?
pd.crosstab(df.Diagnosis, df.Gender)
```

```python
sns.countplot(x = 'Diagnosis', data = df, hue = 'Gender')
plt.title("diabetes frequency for Gender")
plt.xlabel("0 = no Diabetes, 1 = Diabetes");
```

```python
#number of male is more than double in our dataset than female

#more than '45% male' has heart disease and '75% female' has heart disease
```

```python
#let's move to question 3

#3.people of which Gender has which type of Diagnosis most?

#counting values for different Diagnosis
df.Hypertension.value_counts()
```

```python
#plotting a bar chart
df.Hypertension.value_counts().plot(kind = 'bar', color = ['salmon', 'lightskyblue', 'springgreen', 'khaki'])
plt.title('no of people having Hypertension');
```

```python
pd.crosstab(df.Gender,df.Hypertension)
```

```python
pd.crosstab(df.Hypertension, df.Gender).plot(kind = 'bar', color = ['coral', 'lightskyblue', 'plum', 'khaki'])
plt.title('type of Diagnosis for Gender');
plt.xlabel('0 = female, 1 = male');
```

```python
#most of 'male' has 'type 0' chest pain and least of 'male' has 'type 4' pain

#in case of 'female' 'type 0' and 'type 2' percentage is almost same
```
